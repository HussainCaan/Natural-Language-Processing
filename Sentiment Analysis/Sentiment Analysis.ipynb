```python
text = open('cleaned_read.txt', 'r', encoding = 'utf-8').read()
print(text)
```

```python
# Tokenization is the process of breaking down text into smaller units called tokens, which can be words, phrases, symbols, or other meaningful elements. This is a fundamental step in natural language processing (NLP) and text analysis, as it allows for easier manipulation and analysis of the text data.
Tokenized_text = text.split()
Tokenized_text
```

```python
# Stop words removal like "is", "I", "and", "the", "in", etc. are commonly used words that do not carry significant meaning and are often removed from text data during preprocessing to improve the efficiency and effectiveness of natural language processing (NLP) tasks.
stop_words = ["i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself", "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself", "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these", "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do", "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while", "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before", "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again", "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each", "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than", "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"]

final_words = []
for word in Tokenized_text:
    if word not in stop_words:
        final_words.append(word)

final_words
```

# NLP Emotion Algorithm

```python
# Steps to perform NLP Emotion Algorithm will follow here
# 1: To check if the final_words list have words which are present in emotion.txt file
# - open the emotion.txt file
# - read the file line by line
# - for each line, split the line into word and emotion using ':' as separator
# - if the word is present in final_words list, then add the corresponding emotion to an emotion_list

# 2: Count the frequency of each emotion in the emotion_list using a dictionary or Counter from collections module
# 3: Visualize the emotions using a bar graph or pie chart to show the distribution of emotions in the text data

emotion_file = open('emotions.txt', 'r').read()
emotion_list = []
```

```python
emotion_file
```

```python
final_words
```

```python
# Loop through each line and clear it
for line in emotion_file.splitlines():
    clear_line = line.replace("\n", '').replace(",", '').replace("'", '').strip()
    word, emotion = clear_line.split(':')
    # if word in final_words:
    #     emotion_list.append(emotion)
    # print(f"word: {word}  emotion: {emotion}")
    if word in final_words:
        emotion_list.append(emotion)
        print(word, emotion)
```

```python
from collections import Counter
w = Counter(emotion_list)
```

```python
w
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))

plt.bar(w.keys(), w.values())
plt.xlabel('Emotions')
plt.ylabel('Frequency')
plt.title('Emotion Frequency Distribution')
plt.xticks(rotation=45)
plt.show()
```
