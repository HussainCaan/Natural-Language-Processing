```python
# Text Cleaning is the process of removing or correcting unwanted characters, formatting issues, and inconsistencies in text data to prepare it for analysis or processing. This often involves tasks such as removing special characters, converting text to lowercase, eliminating extra spaces, correcting spelling errors, and standardizing formats. Text cleaning is a crucial step in natural language processing (NLP) and data preprocessing to ensure that the text data is clean, consistent, and ready for further analysis or machine learning tasks.
# I love Python! is in the file.

text = open("read.txt", "r", encoding="utf-8").read()
print(text)
```

```python
# Converting to lowercase
text = text.lower()
print("Lowercase:", text)
```

```python
# removing punctuation like !,.?;:'" etc.
import string

cleaned_text = text.translate(str.maketrans('', '', string.punctuation)) 
cleaned_text = cleaned_text.replace('\n', '').strip()
```

```python
cleaned_text
```

```python
with open("cleaned_read.txt", "w", encoding="utf-8") as f:
    f.write(cleaned_text)
```
